```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import time
import uuid

from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense, Dropout, Activation
from tensorflow.keras.models import model_from_json
from keras.preprocessing.image import ImageDataGenerator ## Reference - https://keras.io/preprocessing/image/
```

```python
path = "C:/Users/User/Project---Hand-Gesture-to-Text/Dataset/"
```

## 1) Collect Dataset

```python
## Creating Folder to Store Images
os.mkdir(os.path.join(path, "train"))
os.mkdir(os.path.join(path, "test"))
os.mkdir(os.path.join(path, "captured"))
for i in range(26):
    label = chr(ord('A') + i)
    os.mkdir(os.path.join(path + "train", label))
    os.mkdir(os.path.join(path + "test", label))
    os.mkdir(os.path.join(path + "captured", label))
```

```python
## SideBar
def sidebar(dict):
    for i in range(26):
        label = chr(ord('A') + i)
        cv2.putText(frame, label + " :: " + str(dict[label]), (10, 10 + i * 15), cv2.FONT_HERSHEY_PLAIN, 1, (0, 0, 0), 2)

## Count of Images Present in the Folder
def Count(directory):
    dict = {}
    for i in range(26):
        label = chr(ord('A') + i)
        dict[label] = len(os.listdir(path + directory + "/" + label))
    return dict

## Original Image ==> Gray Scale Image ==> Apply Gaussian Blur Filter
def get_Filtered_Image(frame, size_of_frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5, 5), 2)
    minValue = 70
    th = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)
    ret, output = cv2.threshold(th, minValue, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    output = cv2.resize(output, (size_of_frame, size_of_frame))
    return output
```

```python
## Collecting Images ........
cap = cv2.VideoCapture(0)
while(1):
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)
    
    ## Displaying SideBar
    cnt = Count("captured")
    sidebar(cnt)
    
    ## ROI
    x1 = int(0.5 * frame.shape[1])
    y1 = 10
    x2 = frame.shape[1] - 10
    y2 = int(0.5 * frame.shape[1])
    cv2.rectangle(frame, (220 - 1, 9), (620 + 1, 419), (0, 0, 255), 5)
    ROI = frame[10:410, 220:520]

    ## Original Image ==> Gray Scale Image ==> Apply Gaussian Blur Filter
    output = get_Filtered_Image(ROI, 300)
    cv2.imshow("ROI", output)
    cv2.imshow("Original", frame)

    ## Interrupt Handling
    interrupt = cv2.waitKey(1)
    if(interrupt == 27):
        break
    if(interrupt == -1):
        continue
    label = chr(interrupt - 97 + ord('A'))
    imgName = path + "captured" + "/" + label + "/" + str(cnt[label]) + ".jpg"
    status = cv2.imwrite(imgName, ROI)

cv2.destroyAllWindows()
cap.release()
```

```python
cnt = Count("captured")
print(len(cnt))
print(cnt)
```

## 2) PreProcessing && Spliting of Images into Training and Testing Dataset 

```python
size_of_image = 300
for i in range(26):
    label = chr(ord('A') + i)
    directory = path + "captured" + "/" + label
    for _, _, folder in os.walk(directory):
        ## Splitting into 80-20 % (80% training, 20% testing)
        cnt_for_train = (len(folder) * 75) // 100
        for imgName in folder:
            imgPath = directory + "/" + imgName
            img = cv2.imread(imgPath);
            img = get_Filtered_Image(img, 300)
            if(cnt_for_train >= 0):
                newPath = path + "train" + "/" + label + "/" + imgName
            else:
                newPath = path + "test" + "/" + label + "/" + imgName
            cv2.imwrite(newPath, img)
            cnt_for_train -= 1
```

## 3) Preparing Training & Testing Data

```python
training_Dataset_Path = path + "train"
testing_Dataset_Path = path + "test"


train_batches = ImageDataGenerator().flow_from_directory(training_Dataset_Path,
                                                 target_size = (size_of_image, size_of_image),
                                                 class_mode = "categorical",
                                                 color_mode="grayscale",
                                                 batch_size = 10)
test_batches = ImageDataGenerator().flow_from_directory(testing_Dataset_Path,
                                                 target_size = (size_of_image, size_of_image),
                                                 class_mode = "categorical",
                                                 color_mode="grayscale",
                                                 batch_size = 10)
```

## 4) Visualizing Dataset

```python
train_batches
```

```python
imgs, labels = next(train_batches)
print(imgs[0].shape)
print(len(imgs))
# print(labels)
cnt = 5
for img, label in zip(imgs, labels):
    img = cv2.resize(img, (size_of_image, size_of_image))
    plt.imshow(img, cmap = "gray")
    plt.title(label)
    plt.show()
    cnt -= 1
    if(cnt == 0):
        break
```

## 5) Building Convolution Neural Network Model 

```python
# Initializing the CNN
classifier = Sequential()

# 1st CNN layer and pooling
classifier.add(Conv2D(32, (3, 3), input_shape = (size_of_image, size_of_image, 1), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# 2nd CNN layer and pooling
classifier.add(Conv2D(32, (3, 3), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Flattening the layers
classifier.add(Flatten())

# Adding a fully connected layer
classifier.add(Dense(units=128, activation = 'relu'))
classifier.add(Dropout(0.40))
classifier.add(Dense(units=96, activation = 'relu'))
classifier.add(Dropout(0.40))
classifier.add(Dense(units=64, activation = 'relu'))
classifier.add(Dense(units=27, activation = 'softmax')) # softmax for more than 2 
```

## 6) Compliling our CNN Model

```python
classifier.compile(optimizer='adam', loss = 'categorical_crossentropy', metrics=['accuracy'])
```

```python
classifier.summary()
```

## 6) Training the Model

```python
# steps_per_epoch = total images / batch size
classifier.fit_generator(
        train_batches,
        steps_per_epoch = 13, # No of images in training set
        epochs=5,
        validation_data = test_batches,
        validation_steps = 0) # No of images in test set
```

## 6) Saving Model to JSON File

```python
model_json = classifier.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)
print('========Model Saved========')
classifier.save_weights('weights.h5')
print('========Weights Saved========')
```

## 7) Retrieve Model From JSON File

```python
with open("model.json", "r") as file:
    model_json = file.read()
print("=======Structure Loaded=======")
loaded_model = model_from_json(model_json)
print("=======Weights Loaded=======")
loaded_model.load_weights("weights.h5")
```
